```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')

rv = norm.rvs(loc=22.5, scale=1.25, size=100)
ax.hist(rv, density=True, histtype='stepfilled', alpha=0.5)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
q3 = 1 - norm(loc=22.5, scale=1.25).cdf(28)
print('Answer: {:.7f}'.format(q3))
```
